```python
mean1 = np.mean(data['changes_in_cases'])
mean2 = statistics.mean(data['changes_in_deaths'])
std1 = np.std(data['changes_in_cases'])
std2 = statistics.stdev(data['changes_in_deaths'])
count1 = data['changes_in_cases'].count()
count2 = data['changes_in_deaths'].agg('count')
print(mean1)
print(std1)
print(count1)
print('---')
print(mean2)
print(std2)
print(count2)

boxplot1 = np.random.normal(mean1, std1, count1)
boxplot2 = np.random.normal(mean2, std2, count2)

box_plot = [boxplot1, boxplot2]
bp = plt.boxplot(boxplot1)
plt.style.use('grayscale') #greyscale?
plt.show()

bp = plt.boxplot(boxplot2)
plt.style.use('classic')
plt.show()
```
